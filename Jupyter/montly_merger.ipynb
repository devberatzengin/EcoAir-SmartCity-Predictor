```python
import pandas as pd
import os
import glob

# --- CONFIGURATION ---
DATA_DIR = '/Users/beratzengin/Desktop/Github/EcoAir SmartCity Predictor/Data'             # The folder where your station folders are
OUTPUT_DIR = '/Users/beratzengin/Desktop/Github/EcoAir SmartCity Predictor/Feature Engineering' # Where you want to save Beşiktaş.csv, Aksaray.csv, etc.

# Create output folder if it doesn't exist
if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)

# Get all station folders (excluding files like stations_info.csv)
station_folders = [f.path for f in os.scandir(DATA_DIR) if f.is_dir()]

print(f"Found {len(station_folders)} station folders.")

for folder_path in station_folders:
    station_name = os.path.basename(folder_path)
    
    # Ignore hidden folders like --traffic if necessary
    if station_name.startswith('--'):
        continue
        
    print(f"Merging data for station: {station_name}...")
    
    # 1. Get all monthly CSV files inside the station folder
    csv_files = glob.glob(os.path.join(folder_path, "raw_*.csv"))
    
    if not csv_files:
        print(f"   No monthly files found in {station_name}, skipping.")
        continue
        
    # 2. Read and combine them
    # We use a list comprehension for speed
    temp_list = []
    for file in csv_files:
        df = pd.read_csv(file)
        # Optional: Add a column to keep track of which month/file it came from
        # df['source_file'] = os.path.basename(file)
        temp_list.append(df)
        
    station_df = pd.concat(temp_list, axis=0, ignore_index=True)
    
    # 3. Save with the station name
    output_path = os.path.join(OUTPUT_DIR, f"{station_name}.csv")
    station_df.to_csv(output_path, index=False)
    
    print(f"   Successfully created {station_name}.csv")

print("\n--- ALL DONE! ---")
```

```python
import pandas as pd
import os
import glob

# --- CONFIGURATION ---
DATA_DIR = '/Users/beratzengin/Desktop/Github/EcoAir SmartCity Predictor/Feature Engineering'  # Ana veri klasörün

# Get all station folders
station_folders = [f.path for f in os.scandir(DATA_DIR) if f.is_dir()]

for folder_path in station_folders:
    station_name = os.path.basename(folder_path)
    
    # Gereksiz klasörleri atla
    if station_name.startswith('--'):
        continue
        
    # Sadece aylık ham verileri bul (raw_ ile başlayanlar)
    csv_files = glob.glob(os.path.join(folder_path, "raw_*.csv"))
    
    if not csv_files:
        continue
    
    print(f"Processing {station_name}...")
    
    # Tüm ayları oku ve birleştir
    df_list = [pd.read_csv(file) for file in csv_files]
    station_df = pd.concat(df_list, axis=0, ignore_index=True)
    
    # Kendi klasörüne "İstasyonAdı_RAW.csv" olarak kaydet
    output_filename = f"{station_name}_RAW.csv"
    output_path = os.path.join(folder_path, output_filename)
    
    station_df.to_csv(output_path, index=False)
    print(f"Done! Created: {output_path}")

print("\nAll stations processed successfully!")
```

```python
import os

# --- CONFIGURATION ---
DATA_DIR = '/Users/beratzengin/Desktop/Github/EcoAir SmartCity Predictor/Feature Engineering' # Ana klasörün yolu

# Tüm alt klasörleri ve dosyaları gez (os.walk en garantisidir)
for root, dirs, files in os.walk(DATA_DIR):
    for filename in files:
        # Sadece .csv uzantılı ve zaten sonu _RAW ile bitmeyen dosyaları seç
        if filename.endswith(".csv") and not filename.endswith("_RAW.csv"):
            
            # Dosya adını ve uzantısını ayır (Örn: raw_april ve .csv)
            name, extension = os.path.splitext(filename)
            
            # Yeni dosya adını oluştur (Örn: raw_april_RAW.csv)
            new_filename = f"{name}_RAW{extension}"
            
            # Mevcut yol ve yeni yolu belirle
            old_path = os.path.join(root, filename)
            new_path = os.path.join(root, new_filename)
            
            # Yeniden adlandır
            os.rename(old_path, new_path)
            print(f"Renamed: {filename} -> {new_filename}")

print("\nAll files renamed successfully!")
```
