```python
df.drop(columns=['0'], inplace=True)
```

```python
df = df.sort_values(by=['Station_Name', 'date'])
```

```python
df[['PM10', 'SO2', 'O3', 'NO2', 'CO']] = df[['PM10', 'SO2', 'O3', 'NO2', 'CO']].fillna(method='ffill')
```

```python
df[['PM10', 'SO2', 'O3', 'NO2', 'CO']] = df[['PM10', 'SO2', 'O3', 'NO2', 'CO']].ffill()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['hour'] = df['date'].dt.hour
df['day_of_week'] = df['date'].dt.dayofweek # 0=Pazartesi
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
df['month'] = df['date'].dt.month
```

```python
df['prev_PM10'] = df.groupby('Station_Name')['PM10'].shift(1)
```

```python
df['traffic_rolling_3h'] = df.groupby('Station_Name')['total_vehicles'].transform(lambda x: x.rolling(3).mean())
```

```python
df = pd.get_dummies(df, columns=['Station_Name'], prefix='station')
```
