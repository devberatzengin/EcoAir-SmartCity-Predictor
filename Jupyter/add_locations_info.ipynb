```python
import pandas as pd
import glob
import os
```

```python

FEATURE_DIR ='/Users/beratzengin/Desktop/Github/EcoAir SmartCity Predictor/Feature Engineering'

raw_files = glob.glob(os.path.join(FEATURE_DIR, "*_RAW.csv"))

combined_list = []
```

```python

for file in raw_files:
    df = pd.read_csv(file)
    
    station_name = os.path.basename(file).replace('_RAW.csv', '')
    
    df['Station_Name'] = station_name
    
    combined_list.append(df)

master_df = pd.concat(combined_list, axis=0, ignore_index=True)

master_df.to_csv('MASTER_MERGED_RAW.csv', index=False)
print(f"Bitti kanka! Toplam {len(master_df)} satır veri birleştirildi.")
```
